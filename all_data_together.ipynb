```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
train1 = pd.read_csv('train_pop_per_pvz_by_iso.csv')
train2 = pd.read_csv('train_nearest_neighbor_distance.csv')
train3 = pd.read_csv('train_with_selected_commercial_density.csv')
train4 = pd.read_csv('train_landuse.csv')
train5 = pd.read_csv('train_metro_count_per_pvz.csv')


data = pd.merge(train1, train2, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
data = pd.merge(data, train3, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
data = pd.merge(data, train4, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
data = pd.merge(data, train5, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])

list_idx_train = list(data.id)

data.fclass.fillna('unknown', inplace=True) 
data
```

```python
test1 = pd.read_csv('test_pop_per_pvz_by_iso.csv')
test2 = pd.read_csv('test_nearest_neighbor_distance.csv')
test3 = pd.read_csv('test_with_selected_commercial_density.csv')
test4 = pd.read_csv('test_landuse.csv')
test5 = pd.read_csv('test_metro_count_per_pvz.csv')



test = pd.merge(test1, test2, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
test = pd.merge(test, test3, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
test = pd.merge(test, test4, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])
test = pd.merge(test, test5, on = ['id', 'sales_per_month', 'region', 'locality', 'x', 'y', 'people_iso_3', 'people_iso_5', 'people_iso_10', 'people_iso_15', 'people_iso_30', 'type'])

list_idx_test = list(test.id)


test.fclass.fillna('unknown', inplace=True) 
test
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(pd.concat([test['fclass'], data['fclass']]))
test['fclass_encoded'] =label_encoder.transform(test['fclass'])
data['fclass_encoded'] =label_encoder.transform(data['fclass'])

label_encoder = LabelEncoder()
label_encoder.fit(pd.concat([test['locality'], data['locality']]))
test['locality_encoded'] =label_encoder.transform(test['locality'])
data['locality_encoded'] =label_encoder.transform(data['locality'])

label_encoder = LabelEncoder()
label_encoder.fit(pd.concat([test['region'], data['region']]))
test['region_encoded'] =label_encoder.transform(test['region'])
data['region_encoded'] =label_encoder.transform(data['region'])
```

```python
whole_data = pd.concat([data, test], ignore_index = True)
whole_data
```

```python
#fixing nan regions
no_region_idx = whole_data[whole_data.region.isna()].index
whole_data.at[no_region_idx[0], 'region'] = 'Карагандинская область'
whole_data.at[no_region_idx[1], 'region'] = 'Акмолинская область'
```

```python
stats = pd.read_csv('stats1.csv')
stats.region = stats.region.apply(lambda x: x.replace('_Вайоцдзорская область', 'Вайоцдзорская область'))
```

```python
whole_data = pd.merge(whole_data, stats,on ='region', how='left')
whole_data

```
