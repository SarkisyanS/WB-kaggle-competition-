```python
import pandas as pd
import numpy as np

import osmnx as ox
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
train = pd.read_csv('train.csv')
train['type'] = 'train'
test = pd.read_csv('test.csv')
test['type'] = 'test'
data = pd.concat([train, test], ignore_index = True)
data
```

```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
from tqdm import tqdm

amenity_categories = ['restaurant', 'pub', 'ice_cream', 'food_court', 'fast_food',
                      'cafe', 'biergarten', 'bar', 'atm', 'payment_terminal', 'bank',
                      'bureau_de_change', 'dentist', 'pharmacy', 'veterinary', 'casino',
                      'cinema', 'nightclub', 'stripclub', 'theatre', 'events_venue', 
                      'post_office' ,'marketplace']

amens300 = []
amens500 = []
for index, row in tqdm(data.iterrows()):
    point = (row['y'], row['x'])  # Fetch the point coordinates from each row
    
    try:
        amen_count_300 = ox.features.features_from_point(center_point=point, tags={'amenity': amenity_categories}, dist=300).amenity.count()
        amen_count_500 = ox.features.features_from_point(center_point=point, tags={'amenity': amenity_categories}, dist=500).amenity.count()
        amens300.append(amen_count_300)
        amens500.append(amen_count_500)
        
    except:
        amens300.append(0)
        amens500.append(0)

data['commercial_density_300'] = amens300
data['commercial_density_500'] = amens500


```

```python
commercial_density_300 = amens300
commercial_density_500 = amens500

new_train = data[data['type'] == 'train']
new_test = data[data['type'] == 'test']

new_train.to_csv('train_with_selected_commercial_density.csv', index = False)
new_test.to_csv('test_with_selected_commercial_density.csv', index = False)

#data.to_csv('train_with_selected_commercial_density.csv', index = False)
```
